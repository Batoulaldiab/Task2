* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
import matplotlib.pyplot as plt

# Create a figure object
fig = plt.figure()

# Add an axis to the figure at [0,0,1,1]
ax = fig.add_axes([0, 0, 1, 1])  # This creates the entire plot area

# Sample data (replace with your actual data)
x = [0, 20, 40, 60, 80, 100]
y = [0, 50, 100, 150, 200, 250]


# Plot the data with labels and title
ax.plot(x, y, label='Sample Data')  # Plot the line
ax.set_xlabel('X-axis')  # Set X-axis label
ax.set_ylabel('Y-axis')  # Set Y-axis label
ax.set_title('Sample Plot')  # Set plot title

# Add legend
ax.legend()

# Show the plot
plt.show()
```
